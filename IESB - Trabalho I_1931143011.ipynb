```python
### Deve-se colocar o código aqui
#Fazendo DE => PARA
clusteres = {
    "c1":"0.png",
    "c2":"Z.png",
    "c3":"Y.png",
    "c4":"1.png,2.png,3.png",
    "c5":"4.png,5.png,6.png",
    "c6":"7.png,8.png,9.png",
    "c7":"A.png,B.png,C.png",
    "c8":"D.png,E.png,F.png",
    "c9":"G.png,H.png,I.png",
    "c10":"J.png,K.png,L.png",
    "c11":"M.png,N.png,O.png",
    "c12":"P.png,Q.png,R.png",
    "c13":"S.png,T.png,U.png",
    "c14":"V.png,W.png,X.png"
}

de_para = {}

totalcaracteres = 0;
for arquivo in os.listdir('/content/drive/My Drive/Colab Notebooks/aprendizado-nao-supervisionado/trabalho1/dataset/encontrado/'):
    #encontrando cluster
    for cluster, caracter in clusteres.items():
      vcaracter = caracter.split(",")
      totalelementos = len(vcaracter)
      if (totalelementos == 3):
        for vchar in vcaracter:
          if (vchar == arquivo):
            clusterencontrado = cluster                     
      else:
        if (caracter == arquivo):
          clusterencontrado = cluster
    
    de_para[arquivo] = clusterencontrado
    totalcaracteres += 1
    
print("Total de Caracteres: ",totalcaracteres)
print(de_para)
```

**Segundo exercício:**

Utilize um KNN (algoritmo de vizinhos próximos) para classificar os rótulos que você criou para os clusters do exercício I. Lembrando que deve efetuar a separação de dados para testar e validar as informações. 

**Dicas:**

- Faça um Data Augumentation.
- Faça padding centralizando o conteúdo das imagens para padronizá-las para o mesmo tamanho.

```python
#Fazenda Data Augumentation
import random
from scipy import ndarray
import skimage as sk
from skimage import transform
from skimage import util

def random_rotation(image_array: ndarray):
    # faz uma rotacao a esquerda e a direita de 25%
    random_degree = random.uniform(-25, 25)
    return sk.transform.rotate(image_array, random_degree)

def random_noise(image_array: ndarray):
    # adiciona ruido na imagem
    return sk.util.random_noise(image_array)

arquivogerado = 1;
caminho = '/content/drive/My Drive/Colab Notebooks/aprendizado-nao-supervisionado/trabalho1/dataset/encontrado/'
for arquivo in os.listdir(caminho):
  print("Gerando Data Augumentation: ",arquivo)
  imagembin = sk.io.imread(caminho+arquivo)
  clusterencontrado = de_para[arquivo]
  
  rotacao = 1
  maximorotacoes = 5
  while (rotacao <= maximorotacoes):
    rr = random_rotation(imagembin)
    nomearquivo = '%s_da_rr_%s_%s.png' % (arquivo,arquivogerado,rotacao)
    novoarquivorr = caminho+nomearquivo
    
    if (os.path.exists(novoarquivorr)):
      os.remove(novoarquivorr)
      
    sk.io.imsave(novoarquivorr,rr)
    de_para[nomearquivo] = clusterencontrado
    rotacao += 1
  
  rn = random_noise(imagembin)
  nomearquivo = '%s_da_rn_%s.png' % (arquivo,arquivogerado)
  novoarquivorn = caminho+nomearquivo
  
  if (os.path.exists(novoarquivorn)):
      os.remove(novoarquivorn)
      
  sk.io.imsave(novoarquivorn,rn)
  de_para[nomearquivo] = clusterencontrado
  
  arquivogerado += 1
  
```

```python
#Preparando vetorizacao das imagens
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from imutils import paths
import numpy as np
import argparse
import imutils
import cv2
import os

def image_to_feature_vector(image, size=(32, 32)):
	return cv2.resize(image, size).flatten()

def extract_color_histogram(image,bins=(8,8,8)):
  hsv = cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
  hist = cv2.calcHist([hsv],[0,1,2],None,bins,[0,180,0,256,0,256])
  
  if (imutils.is_cv2()):
	  hist=cv2.normalize(hist)
  else:
    cv2.normalize(hist,hist)

  return hist.flatten()

#Encontrando o cluster via KNN
linhasImagens = []
caracteristicas = []
labels = []
caminho = '/content/drive/My Drive/Colab Notebooks/aprendizado-nao-supervisionado/trabalho1/dataset/encontrado/'

for arquivo in os.listdir(caminho):
  imagembin = cv2.imread(caminho+arquivo)
  
  pixels = image_to_feature_vector(imagembin)
  histograma = extract_color_histogram(imagembin)
  linhasImagens.append(pixels)
  caracteristicas.append(histograma)
  labels.append(de_para[arquivo])
  
```

```python
#Fazendo KNN
(trainLinhas, testLinhas, trainLinhaLabels, testLinhaLabels) = train_test_split(linhasImagens, labels, test_size=0.25, random_state=42)
(trainCaracteristicas, testCaracteristicas, trainLabels, testLabels) = train_test_split(caracteristicas, labels, test_size=0.25, random_state=42)

#Para "linhas"
modelLinhas = KNeighborsClassifier(n_neighbors=7,n_jobs=-1)
modelLinhas.fit(trainLinhas, trainLinhaLabels)
accLinhas = modelLinhas.score(testLinhas, testLinhaLabels)
print("Para Linhas: ",accLinhas)

#Para "histogramas"
modelHist = KNeighborsClassifier(n_neighbors=7,n_jobs=-1)
modelHist.fit(trainCaracteristicas, trainLabels)
accHist = modelHist.score(testCaracteristicas, testLabels)
print("Para Histograma: ",accHist)
```
